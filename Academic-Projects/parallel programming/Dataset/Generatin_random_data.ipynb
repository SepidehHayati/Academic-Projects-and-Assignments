```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
center_1 = np.array([1,1])
center_2 = np.array([3000,3000])
center_3 = np.array([5000,1])

# standard deviation
sd = 400

# random data per each point
data_1 = sd * np.random.randn(36000, 2) + center_1
data_2 = sd * np.random.randn(36000, 2) + center_2
data_3 = sd * np.random.randn(48000, 2) + center_3

data = np.concatenate((data_1, data_2, data_3), axis = 0)

# scattering
plt.scatter(data[:,0], data[:,1], s=7)
```

```python
print(len(data))
```
